```python
# Region Ã— Overall Effectiveness
region_effectiveness = (
    df.groupby('region')['overall_effectiveness']
      .value_counts()
      .unstack(fill_value=0)
)

# Improvement / decline
change_counts = df['improved/_declined/_stayed_the_same'].value_counts()

# Provider type Ã— Overall Effectiveness
provider_effectiveness = (
    df.groupby('provider_type')['overall_effectiveness']
      .value_counts()
      .unstack(fill_value=0)
)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle(
    "EDA Dashboard: Further Education & Skills Ofsted Inspections (2024â€“2025)",
    fontsize=18,
    fontweight="bold"
)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 1. Overall Effectiveness
df['overall_effectiveness'].value_counts().plot(
    kind='bar',
    ax=axes[0, 0]
)
axes[0, 0].set_title("Overall Effectiveness Distribution")
axes[0, 0].set_xlabel("Rating")
axes[0, 0].set_ylabel("Number of Providers")
axes[0, 0].tick_params(axis='x', rotation=45)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 2. Region vs Overall Effectiveness (stacked)
region_effectiveness.plot(
    kind='bar',
    stacked=True,
    ax=axes[0, 1]
)
axes[0, 1].set_title("Overall Effectiveness by Region")
axes[0, 1].set_xlabel("Region")
axes[0, 1].set_ylabel("Inspection Count")
axes[0, 1].legend(title="Rating", fontsize=8)
axes[0, 1].tick_params(axis='x', rotation=45)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 3. Improvement vs Decline
change_counts.plot(
    kind='bar',
    ax=axes[1, 0]
)
axes[1, 0].set_title("Change Since Previous Inspection")
axes[1, 0].set_xlabel("Inspection Outcome")
axes[1, 0].set_ylabel("Number of Providers")
axes[1, 0].tick_params(axis='x', rotation=45)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 4. Provider Type vs Effectiveness (heatmap)
sns.heatmap(
    provider_effectiveness,
    annot=True,
    fmt="d",
    cmap="Blues",
    ax=axes[1, 1]
)
axes[1, 1].set_title("Provider Type vs Overall Effectiveness")
axes[1, 1].set_xlabel("Overall Effectiveness")
axes[1, 1].set_ylabel("Provider Type")



# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 5. Region vs Overall Effectiveness (heatmap)
sns.heatmap(
    region_effectiveness,
    annot=True,
    fmt="d",
    cmap="Blues",
    ax=axes[2, 0]
)

axes[2, 0].set_title("Region vs Overall Effectiveness")
axes[2, 0].set_xlabel("Overall Effectiveness")
axes[2, 0].set_ylabel("Region")


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 5.Monthly analysis with overall_effectiveness
monthly_trend = df.groupby('month_visit')['overall_effectiveness'].sum()

monthly_trend.plot(kind='pie',ax=axes[2, 1])
axes[2, 1].set_title("Monthly analysis with overall_effectiveness (2024â€“2025)")






plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
